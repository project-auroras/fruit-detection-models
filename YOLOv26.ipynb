```python
from ultralytics import YOLO

# Cargar el modelo preentrenado
model = YOLO('yolo26m.pt')
```

```python
# Entrenar
model.train(
    data="data.yaml",
    epochs=200,
    patience=10,
    imgsz=640,
    batch=8,
)
```

```python
# Evaluar sobre el dataset
metrics = model.val(data="data.yaml", plots=True)

# Mostrar resultados
print(metrics)
```

```python
print(f"Speed: {metrics.speed}")
```

```python
model.benchmark(data='data.yaml', imgsz=640, device='0') # device='0' for GPU
```
