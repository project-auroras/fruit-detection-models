```python
# Load a COCO-pretrained RT-DETR-l model
model = RTDETR("rtdetr-l.pt")

# Display model information (optional)
model.info()
```

```python
# Entrenar
model.train(
    data="data.yaml",
    epochs=200,
    patience=10,
    imgsz=640,
    batch=2,
)
```

```python
#Evaluamos el modelo entrenado
metrics = model.val(data="data.yaml", plots=True)

# Mostrar resultados
print(metrics)
```
