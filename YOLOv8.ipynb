```python
from ultralytics import YOLO

# Cargar el modelo preentrenado
model = YOLO('yolov8l.pt')
```

```python
# Entrenar
model.train(
    data="/home/lidia/Code/vision_models/yolo/Mencia/DatasetMarzo2026/data.yaml",
    epochs=200,
    patience=10,
    imgsz=640,
    batch=8,
)
```

```python
# Evaluar sobre el dataset
metrics = model.val(data="data.yaml", plots=True)

# Mostrar resultados 
print(metrics)
```

```python
from ultralytics.utils.benchmarks import benchmark

# Benchmark en GPU (requiere Nvidia)
benchmark(model="runs/detect/train/weights/best.pt", data="data.yaml", imgsz=640, device=0)
```
