```python
from ultralytics import YOLO

# Cargar el modelo preentrenado
model = YOLO('yolo11m.pt')
```

```python
# Entrenar
model.train(
    data="data.yaml",
    epochs=100,
    patience=5,
    imgsz=512,
    batch=8,
)
```

```python
# Evaluar sobre el dataset
metrics = model.val(data="data.yaml", plots=True)

# Mostrar resultados 
print(metrics)
```

```python
print(f"Speed: {metrics.speed}") # Displays prep, inference, and post-process times
```

```python
from ultralytics.utils.benchmarks import benchmark

# Benchmark en GPU (requiere Nvidia)
benchmark(model="detect/train/weights/best.pt", data="data.yaml", imgsz=640, device=0)
```
